# Concrete Flexural Resistance  - $M_f$ given

For simple checks on concrete flexural resistance

**Note:** Graphic above shows a cantilever of 2'3. The actual site conditions was a cantilever of 4'. The moment generated by 4' cantilever is represented in the calculation below.

```python
## Inputs
M_f = 26*kip*ft
h_c = 8*inch
b = (3 + 7/12) * ft
cover = 1*inch

phi_c = 0.65
f_c = 30*MPa
f_y = 400*MPa
```

## Flexural Resistance

```python
%%render
# Long
d = min(0.9*(h_c), h_c-cover-20*mm/2)
```

```python
def sqrt(x):
    if hasattr(x, "dimensions") and x.dimensions == (1, -1, -2, 0, 0, 0, 0):
        return float(x)**0.5 * MPa
    return x**0.5
```

```python
%%render
A_s_req_400 = 0.0015/MPa*f_c*b*(d - sqrt(d**2 - (3.85*M_f)/(f_c * b)))
```

```python
%%render
A_s_req_300 = 400 / 300 * A_s_req_400
```
