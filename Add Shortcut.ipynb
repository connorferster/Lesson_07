# Add a keyboard shortcut for "Run all"

1. Open the `Advanced Settings Editor`



2. Click `Keyboard Shortcuts`



3. Copy and paste the below into the `User Preferences` panel on the far right:

```json
{"shortcuts": 
    [
        {
            "command": "notebook:run-all-cells",
            "keys": ["Alt R"],
            "selector": "body",
            "disabled": false,
        }
    ]
}
```



4. Close the `Settings` tab

You can now use the keyboard shortcut `Alt r` in any notebook to run all the cells from top to bottom.
